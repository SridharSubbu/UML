Calculating Various Distance Measures
1. Euclidean
2. Manhatten
3. Minkowski
4. Hamming
5. Cosine Similarity


```python
from math import sqrt

#calculate euclidean distance
def euclidean_distance(a,b):
  return sqrt(sum((e1-e2)**2 for e1,e2 in zip(a,b)))

#define data

row1 = [10,20,15,10,5]  #[10,20,30,50,40]
row2 = [12,24,18,8,7] #[12,24,36,48,60]

dist = euclidean_distance(row1,row2)
print(dist)
```

We can perform the same calculation using the euclidean() function from SciPy


```python

from scipy.spatial.distance import euclidean

# Define the two points as numpy arrays
point1 = [10,20,15,10,5]
point2 = [12,24,18,8,7]

# Calculate the Euclidean distance
distance = euclidean(point1, point2)

print(f"The Euclidean distance between point1 and point2 is: {distance}")
```

ManHattan distance between Vectors


```python
from math import sqrt

def manhattan_distance(a,b):
  return sum(abs(e1-e2) for e1,e2 in zip(a,b))

# Define Date
row1 = [10,20,15,10,5]  #[10,20,30,50,40]
row2 = [12,24,18,8,7] #[12,24,36,48,60]

dist = manhattan_distance(row1,row2)
print(dist)
```

Using Scipy Manhattan


```python
# cityblock,

from scipy.spatial.distance import cityblock

# Define the two points as numpy arrays
point1 = [10,20,15,10,5]
point2 = [12,24,18,8,7]

# Calculate the Euclidean distance
distance = cityblock(point1, point2)

print(f"The manhattan distance between point1 and point2 is: {distance}")
```

Minkowski Distance


```python
from math import sqrt

def minkowski_distance(a,b,p):
  return sum(abs(e1-e2)**p for e1, e2 in zip(a,b))**(1/p)

row1 = [10,20,15,10,5]  #[10,20,30,50,40]
row2 = [12,24,18,8,7] #[12,24,36,48,60]


# p=1

dist = minkowski_distance(row1,row2,1)
print(dist)

# p=2

dist = minkowski_distance(row1,row2,2)
print(dist)
```

```python
# Minkowski with SciPy
# cityblock,

from scipy.spatial.distance import minkowski

# Define the two points as numpy arrays
point1 = [10,20,15,10,5]
point2 = [12,24,18,8,7]

# Calculate the Euclidean distance
distance = minkowski(point1, point2,1)

print(f"The minkowski distance between point1 and point2 is: {distance}")

distance = minkowski(point1, point2,2)

print(f"The minkowski distance between point1 and point2 is: {distance}")
```

Hamming Distance


```python
from math import sqrt

def hamming_distance(a,b):
  return sum(abs(e1-e2) for e1,e2 in zip(a,b)) / len(a)

# Define Date
row1 = [0,0,0,0,0,1]
row2 = [0,0,0,0,1,0] #[12,24,36,48,60]

dist = hamming_distance(row1,row2)
print(dist)
```

```python
# Scipy

from scipy.spatial.distance import hamming

# Define the two points as numpy arrays
point1 = [0,0,0,0,0,1]
point2 = [0,0,0,0,1,0]

# Calculate the Euclidean distance
distance = hamming(point1, point2)

print(f"The hamming distance between point1 and point2 is: {distance}")

```

Cosine Similarity

```python
import numpy as np
from numpy.linalg import norm

# Define 2 lists or array

A = np.array([2,1,2,3,2,9])
b = np.array([3,4,2,4,5,5])

print("A: ", A)
print("b: ", b)

cosine = np.dot(A,b)/(norm(A)*norm(b))
print("Cosine Similarity", cosine)
```
